```python
def get_company_info(name, url):
    driver = webdriver.Chrome('C:/Program Files (x86)/Google/Chrome/Application/chromedriver.exe', options=chrome_options)
    wait = WebDriverWait(driver, 2)
    driver.get(url)
    try:
        if driver.find_element_by_xpath('.//div[@data-test="employerOverviewModule"]'):
            # comany_buttom = driver.find_element_by_xpath('.//div[@data-tab-type="overview"]').click()
            sleep(1)
            try:
                if driver.find_element_by_xpath('.//div[@data-test="employerOverviewModule"]//li/div[@data-test="employer-headquarters"]'):
                    headquarter = driver.find_element_by_xpath('.//div[@data-test="employerOverviewModule"]//li/div[@data-test="employer-headquarters"]').text
                else:
                    headquarter = ''
            except:
                headquarter = ''

            try:
                if driver.find_element_by_xpath('.//div[@data-test="employerOverviewModule"]//li/div[@data-test="employer-size"]'):
                    size = driver.find_element_by_xpath('.//div[@data-test="employerOverviewModule"]//li/div[@data-test="employer-size"]').text
                else:
                    size = ''
            except:
                size = ''

            try:
                if driver.find_element_by_xpath('.//div[@data-test="employerOverviewModule"]//li/div[@data-test="employer-founded"]'):
                    founded = driver.find_element_by_xpath('.//div[@data-test="employerOverviewModule"]//li/div[@data-test="employer-founded"]').text
                else:
                    founded = ''
            except:
                founded = ''

            try:
                if driver.find_element_by_xpath('.//div[@data-test="employerOverviewModule"]//li/div[@data-test="employer-type"]'):
                    company_type = driver.find_element_by_xpath('.//div[@data-test="employerOverviewModule"]//li/div[@data-test="employer-type"]').text
                else:
                    company_type = ''
            except:
                company_type = ''

            try:
                if driver.find_element_by_xpath('.//div[@data-test="employerOverviewModule"]//li/a[@data-test="employer-industry"]'):
                    industry = driver.find_element_by_xpath('.//div[@data-test="employerOverviewModule"]//li/a[@data-test="employer-industry"]').text
                else:
                    industry = ''
            except:
                industry = ''

            # try:
            # 	if driver.find_element_by_xpath('.//div[@id="InfoFields"]//span[@id="primaryIndustry.sectorName"]'):
            # 		sector = driver.find_element_by_xpath('.//div[@id="InfoFields"]//span[@id="primaryIndustry.sectorName"]').text
            # 	else:
            # 		sector = ''
            # except:
            # 	sector = ''

            try:
                if driver.find_element_by_xpath('.//span[@data-test="employerDescription"]/button'):
                    button = driver.find_element_by_xpath('.//span[@data-test="employerDescription"]/button').click()
                sleep(0.5)
            except:
                sleep(0.5)

            try:
                if driver.find_element_by_xpath('.//span[@data-test="employerDescription"]'):
                    description = driver.find_element_by_xpath('.//span[@data-test="employerDescription"]').text
                else:
                    description = ''
            except:
                description = ''

            try:
                if driver.find_element_by_xpath('.//div[@data-test="employerOverviewModule"]//li/div[@data-test="employer-revenue"]'):
                    revenue = driver.find_element_by_xpath('.//div[@data-test="employerOverviewModule"]//li/div[@data-test="employer-revenue"]').text
                else:
                    revenue = ''
            except:
                revenue = ''

            try:
                if driver.find_element_by_xpath('.//div[@data-test="employerOverviewModule"]//li/a[@data-test="employer-website"]'):
                    company_url = driver.find_element_by_xpath('.//div[@data-test="employerOverviewModule"]//li/a[@data-test="employer-website"]').get_attribute('href')
                else:
                    company_url = ''
            except:
                company_url = ''
            
            try:
                if driver.find_element_by_xpath('.//div[@class="logo cell"]//img[@alt=" Logo"]'):
                    company_logo = driver.find_element_by_xpath('.//div[@class="logo cell"]//img[@alt=" Logo"]').get_attribute("src")
                else:
                    company_logo = ''
            except:
                company_logo = ''

            res = [name, url, headquarter, size, founded, company_type, industry, description, revenue, company_url, company_logo]

    except:
        res = [name, url] + [''] * 9
    print([name, url])
    driver.close()
    return res
```
